```python
#Load Pandas and Numpy for manipulating numbers and dataframes
#Load Matplotlib and Seaborn for Data Visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the dataset
data = pd.read_csv('Wine Quality Dataset.csv')
```

```python
#Get the columns of the dataframe
data.columns
```

```python
#Get the basic info of the dataset
data.info()
```

```python
#Create a histogram chart of 'fixed acidity' feature
plt.figure(figsize=(11,6))
sns.histplot(data=data, x='fixed acidity', color='orange',edgecolor='linen', alpha=0.5, bins=5)
plt.title='Histogram for fixed acidity'
plt.xlabel='fixed acidity'
plt.ylabel='count'
plt.show()
```

```python
# calculate the mean of 'fixed acidity'
round(data['fixed acidity'].mean(),2)
```

```python
# calculate the median of 'fixed acidity'
data['fixed acidity'].median()
```

```python
# Create a histogram of the 'fixed acidity' feature and also show the mean and median
plt.figure(figsize=(11,6))
sns.histplot(data=data, x='fixed acidity', color='orange', edgecolor='linen',alpha=0.5, bins=5)
#plt.title('Histogram of fixed acidity')
#plt.xlabel('Fixed Acidity')
#plt.ylabel('Count')
plt.vlines(data['fixed acidity'].mean(), ymin=0, ymax=4000, color='blue', label='Mean')
plt.vlines(data['fixed acidity'].median(), ymin=0, ymax=4000, color='red', label='Median')
plt.legend()
plt.show()
```

```python
# Create a histogram of 'volatile acidity' feature
plt.figure(figsize=(11,6))
sns.histplot(data=data, x='volatile acidity', color='green', edgecolor='linen', alpha=0.5, bins=5)
#plt.title('Histogram for volatile acidity')
#plt.xlabel('Volatile Acidity')
#plt.ylabel('Count')
plt.show()
```

```python
# Plot distplot using 'volatile acidity' feature
plt.figure(figsize=(11,6))
sns.distplot(data['volatile acidity'], color='blue')
#plt.xlabel('Volatile acidity')
#plt.ylabel('Density')
plt.show()
```

```python
# Create a histogram for 'citric acid' feature
plt.figure(figsize=(11,6))
sns.histplot(data=data, x='citric acid', color='hotpink', edgecolor='linen', alpha=0.5, bins=5)
#plt.title('Histogram for Citric Acid')
plt.show()
```

```python
# Calculate the mean of 'citric acid'
round(data['citric acid'].mean(),2)
```

```python
# Create a histogram for 'citric acid' and also show its mean and median
plt.figure(figsize=(11,6))
sns.histplot(data=data, x='citric acid',color='hotpink', alpha=0.5, bins=5)
#plt.xlabel='Citric Acid'
#plt.ylabel='density'
plt.vlines(data['citric acid'].mean(), ymin=0, ymax=4000, color='blue', label='Mean')
plt.vlines(data['citric acid'].median(), ymin=0, ymax=4000, color='red', label='Median')
plt.legend()
#plt.title('Histogram for Citric Acid')
```

```python
# Calculate the distplot using the 'citric acid' feature
plt.figure(figsize=(11,6))
sns.distplot(data['citric acid'], color='hotpink')
#plt.title('Distplot of citric acid')
#plt.xlabel('Citric Acid')
#plt.ylabel('Density')
plt.show()
```

```python
# Create a count plot of 'quality' feature
plt.figure(figsize=(11,6))
#sns.barplot(data=data, x='quality', color='hotpink')
sns.countplot(x = 'quality', data =data)
#plt.title('Count Plot for Quality')
plt.show()
```

```python
# count the number of occurences of each quality feature
data['quality'].value_counts()
```

```python
# Calculate the mode of the 'quality' feature
data['quality'].value_counts().index[0]
```

```python
# Create a new pandas series called 'rep_acid', that contain the details of the representative quality of different types of acids
rep_acid = pd.Series(index=['fixed quality', 'volatile acidity', 'citric acidity', 'quality'],
                     data=[data['fixed acidity'].mean(),
                           data['volatile acidity'].mean(),
                           data['citric acid'].mean(),
                     data['quality'].value_counts().index[0]])
```
